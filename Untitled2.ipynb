**BINARY CONVERSION**

```python
varlist =  ['Gender']
def binary_map(x):
    return x.map({'Male': 1, "Female": 0})

dataset[varlist] = dataset[varlist].apply(binary_map)

dataset
```

**CATEGORIAL CONVERSION**

```python

status = pd.get_dummies(dataset['Social Media'], drop_first = True)


dataset = pd.concat([dataset, status], axis = 1)


dataset.drop(['Social Media'], axis = 1, inplace = True)

dataset
```

**OUTLIERS DETECTION**

```python
def detect_outliers_zscore(data):
    outliers = []
    thres = 1
    mean = np.mean(data)
    std = np.std(data)
  
    for i in data:
    
        z_score = (i-mean)/std
        if (np.abs(z_score) > thres):
            outliers.append(i)
    return outliers
```

```python
sample_outliers_Age = detect_outliers_zscore(dataset['Age']) 
print("Outliers in Age column from Z-scores method: ", sample_outliers_Age)
```
